```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

### Experiment setup:
1. Each iteration consists of ~10,000 timesteps.
2. 3 experiments were run by scaling weights in each layer by different factors, i.e., 1(normal), 500, 1000.
3. The value network consists of 2 hidden layer, and the optimisation procedure used was LBFGS.

### Statistics Analyzed:
1. Episodes Reward Mean - The mean of rewards obtained in each iteration which consists of ~10,000 steps in the environment.
2. KL loss-
    a. Before policy update
    b. After policy update
3. Policy entropy-
    a. Before policy update
    b. After policy update
4. Gradient Norm statistics
    a. Mean Norm-2 of weight gradient at each layer
    b. Variance Norm-2 of weight gradient at each layer
5. Gradient Norm Differences between timesteps-
    At each layer, the mean of (||g{t+1}||-||g{t}||)/||g{t}||
    , where ||.|| is L^2-Norm.
6. Variance of activations at each layer.

### Weights scaled by 1000.

```python
%run run_pg.py --env CartPole-v0 --agent modular_rl.agentzoo.TrpoAgent --n_iter=500 --exp_name=1000 --wt_scale=1000
```

## Observations:
When the weights are scaled by 1000, TRPO is not able to converge to an optimal policy within the given iterations. Though when scaled by 500, it is still able to converge to an optimal policy.

### Neural Net Statistics observed:

1. KL_loss and policy entropy show similar trends even when the weights are scaled. 
In case, when weights are scaled by 1000 it may not explore as much to find an optimal policy.

2. The mean norm of gradient at each layer is initially greater for large scaled weights, but tends to decrease with iterations.

3. The difference in gradient norm between time-steps((||g{t+1}||-||g{t}||)/||g{t}||) decreases with iterations as the weights are scaled with a higher value. Overall the values are higher compared to unscaled weights.

4. From the above graphs, variation across activations in each layer shows an increase in the normal setup, but a similar trend is not observed when the weights are scaled by huge numbers.
